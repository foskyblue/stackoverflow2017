```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import Utils as u
%matplotlib inline

# Data gathering
df = pd.read_csv('./survey_results_public.csv')
schema = pd.read_csv('./survey_results_schema.csv')
df.head()
```

```python
df_temp = df

# Group by WebDeveloperType, get the mean based on JobSatisfaction
df_temp.groupby('WebDeveloperType').mean()['JobSatisfaction']
```

```python
# Utilize the get_unique function to find all true unique values and display a bar plot.
method_vals = u.get_unique(df['WebDeveloperType'])
methodology = u.clean_and_plot(df, method_vals, 'WebDeveloperType', 'web developer', title='Web Developer Types')
```

```python
# Get the mean of each WebDeveloperType

fe_web_dev = df[(df['WebDeveloperType'] == 'Front-end Web developer') & (df['Salary'].isnull() == False)]['Salary'].mean()
fs_web_dev = df[(df['WebDeveloperType'] == 'Full stack Web developer') & (df['Salary'].isnull() == False)]['Salary'].mean()
be_web_dev = df[(df['WebDeveloperType'] == 'Back-end Web developer') & (df['Salary'].isnull() == False)]['Salary'].mean()

print('Average Front-end Web developer Salary is $',fe_web_dev)
print('Average Full stack Web developer Salary is $',fs_web_dev)
print('Average Back-end Web developer is $',be_web_dev)
```

```python
# Create a Pandas Series that can be used to easliy get a bar plot.
df_dev = pd.Series({'Front-end Web developer': fe_web_dev, 'Full stack Web developer': fs_web_dev, 'Back-end Web developer': be_web_dev})

df_dev.plot(kind="bar");
plt.title('Web developers average salary');
```
